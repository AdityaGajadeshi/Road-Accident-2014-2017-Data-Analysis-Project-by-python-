```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as py
```

```python
dataset = pd.read_csv("C:\\Users\\SSB\\OneDrive\\Documents\\Project_Road_Accident.csv")
```

```python
df = pd.DataFrame(dataset)
#df_cleaned = df.dropna()
df.replace(0, pd.NA, inplace=True)
df_cleaned = df.dropna()
df_cleaned.to_csv('cleaneddataset.csv',index=False)
```

```python
df_cleaned.isnull().sum()
```

```python

x=df_cleaned['killed during 2014-17']
y=df_cleaned['States/UTs']
plt.figure(figsize=(8,8))
plt.xlabel('killed during 2014-17')
plt.ylabel('States/UTs')
plt.plot(x,y,marker='o',color='orange')
plt.title("killed during 2014-17")
plt.show()
```

```python
y=df_cleaned['Injured 2014_17']
x=df_cleaned['States/UTs']
plt.figure(figsize=(10,10))
plt.barh(x,y,color='red')
plt.ylabel('Injured 2014_17')
plt.xlabel('States/UTs')
plt.title("INJURED IN 2014-17")
#plt.barh(x,y,marker='o',color='red')
plt.show()
```

```python
x=df_cleaned['States/UTs']
y=df_cleaned['Accident SH2014-17']
plt.figure(figsize=(15,8))
plt.barh(x,y,color='yellow')
plt.xlabel('States/UTs')
plt.ylabel('Accident SH2014-17')
plt.title("ACCIDENT IN STATE HIGHWAY DURING 2014-17")
plt.show()
```

```python
x=df_cleaned['Accident NH 2014-17']
y=df_cleaned['States/UTs']
plt.figure(figsize=(20,20))
plt.xlabel('Accident NH 2014-17')
plt.ylabel('States/UTs')
plt.pie(x,labels=y,autopct='%1.1f%%')
plt.title("ACCIDENT IN NATIONAL HIGHWAY DURING 2014-17")
plt.show()
```
